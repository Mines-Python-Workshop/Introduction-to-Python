# Example: Pair Distribution Function plotting

Uses data from "PDF_data" folder

```python
# import PDF plotting package from py_figures

from py_figures import PDF
```

```python
# CHANGE ACCORDING TO WHERE EXAMPLE DATA IS SAVED

data_path = "C:/Users/sincl/packages/py_figures/py_figures/PDF_data/"
```

```python
# import data

r, G, Gdiff, Gcalc = PDF.import_PDF(data_path, "Li3InCl6_cell3_no_mixing_refinement.txt", 4)
```

```python
title = r"Li$_3$InCl$_6$, No site mixing"
x_ax = (1, 22)
y_ax = (-1.25, 2)

plt = PDF.plot_PDF(r, G, Gcalc, Gdiff=Gdiff, fit_color="#FA8072", x_lim=x_ax, y_lim=y_ax)

plt.text(1.2, 1.95, title, ha="left", va="top", fontsize="16")
```
