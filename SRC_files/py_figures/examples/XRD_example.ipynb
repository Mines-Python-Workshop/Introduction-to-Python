# XRD plotting examples

Uses data from "XRD_data" folder

```python
# import XRD plotting package from py_figures

import sys
sys.path.append("C:/Users/sincl/packages/py_figures/py_figures/")


import XRD
```

```python
# CHANGE ACCORDING TO WHERE EXAMPLE DATA IS SAVED

data_path = "C:/Users/sincl/packages/py_figures/py_figures/XRD_data/"
```

```python
# import data

expt_tt, expt_obs, expt_err, expt_calc, expt_diff = XRD.import_file(data_path, "SRC001-020_no_adj.txt", 2)
p0_q, p0_ints = XRD.import_file(data_path, "n1000_p0.txt", 0)
p10_q, p10_ints = XRD.import_file(data_path, "n1000_p10.txt", 0)
p20_q, p20_ints = XRD.import_file(data_path, "n1000_p20.txt", 0)
p30_q, p30_ints = XRD.import_file(data_path, "n1000_p30.txt", 0)

h, k, l, hkl_tt, hkl_ints = XRD.import_file(data_path, "Li3ScCl6_hkl.txt", 3)

wl = 0.459744
expt_q = XRD.tt_to_q(expt_tt, wl)
hkl_q = XRD.tt_to_q(hkl_tt, wl)
```

# Plot a single XRD refinement

```python
# plot experimental data, calculated fit, and difference curve
ax1 = XRD.single_fit(expt_q, expt_obs, expt_calc, expt_diff, (0.9, 3), (-18000, 30000), isQ=True, Q_wl=wl)

# add title
title = r"Li$_3$ScCl$_6$, 11-BM-B"
ax1.text(0.92, 29500, title, ha="left", va="top", fontsize="16")

# add (hkl) tick marks
XRD.add_hkl(ax1, hkl_q, 1500, 0.005, -17000, "#97DB4F")
hkl_txt = "Theoretical"
ax1.text(2.98, -13500, hkl_txt, ha="right", va="top", fontsize="14", color="#97DB4F")
```

# Plot refinement, (hkl) ticks, and difference curve on separate subplots

```python
# plot data
ax2 = XRD.hkl_diff_subplots(expt_q, expt_obs, expt_calc, expt_diff, hkl_q, (0.9, 3.2), (-1000, 30000), (-5000, 14000), Q_wl=wl)

# add title
title = r"Li$_3$ScCl$_6$, 11-BM-B"
ax2[0].text(0.92, 29500, title, ha="left", va="top", fontsize="16")

# add hkl label
hkl_txt = "Theoretical"
ax2[1].text(3.18, 0.5, hkl_txt, ha="right", va="center", fontsize="14", color="#97DB4F")
```

# Plot multiple datasets on one plot

```python
# create lists of Q and intensity data sets
q_data = [p0_q, p10_q, p20_q, p30_q]
ints_data = [p0_ints, p10_ints, p20_ints, p30_ints]

# normalize intensities
for i in range(len(ints_data)):
    ints_data[i] = XRD.norm(ints_data[i])
    
# label list
l = ["Unfaulted", r"$P=10\%$", r"$P=20\%$", r"$P=30\%$"]
```

```python
# plot data
ax3 = XRD.stacked_single_plot((0.9, 2.4), (-0.1, 2.5), 4, q_data, ints_data, 0.6, labels=l, label_offsets=(0.01, 0.2), 
                              Q_wl=wl, isNorm=True)

# add title
title = r"Li$_3$ScCl$_6$, $N=1000$, Simulated"
ax3.text(0.92, 2.45, title, ha="left", va="top", fontsize="16")
```

# Plot multiple datasets in separate subplots

```python
# plot data
ax4 = XRD.stacked_subplots((0.9, 2.4), (-0.1, 0.6), 4, q_data, ints_data, labels=l, label_offsets=(0.01, 0.2), Q_wl=wl, isNorm=True)

# add title
title = r"Li$_3$ScCl$_6$, $N=1000$, Simulated"
ax4[0].text(0.92, 0.55, title, ha="left", va="top", fontsize="16")
```
