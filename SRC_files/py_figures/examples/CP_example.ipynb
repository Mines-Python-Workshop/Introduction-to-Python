# Example: Chronopotentiometry plotting

Uses data from "CP_data" folder

```python
# import CP plotting package from py_figures

from py_figures import CP
```

```python
# CHANGE ACCORDING TO WHERE EXAMPLE DATA IS SAVED

data_path = "C:/Users/sincl/packages/py_figures/py_figures/CP_data/"

# import .mpr data files
pos_fn = "In7Zr3_cell5_100uA_02_CP_C01"
neg_fn = "In7Zr3_cell5_100uA_04_CP_C01"

full_pos_CP, full_neg_CP = CP.import_biologic_CP(pos_fn, neg_fn, data_path)
```

```python
# separate CP data by cycle

num_pts_per_cycle = 1201

pos_CP, neg_CP, num_pos, num_neg = CP.sep_CP_cycles(full_pos_CP, full_neg_CP, num_pts_per_cycle)
```

```python
title = r"Li$_{2.7}$In$_{0.7}$Zr$_{0.3}$Cl$_6$"

# plot data
ax = CP.plot_CP(pos_CP, neg_CP, num_pos, x_lim=(-1500, 40500), y_lim=(-0.325, 0.675))

# add title
ax.text(-1000, 0.655, title, fontsize="16", ha="left", va="top")

# add experiment info
expt_info = CP.expt_info(100, num_pos)
ax.text(40000, 0.655, expt_info, fontsize="14", ha="right", va="top")
```
