# Example: EIS plotting

Uses data from "EIS_data" folder

```python
# import EIS plotting package from py_figures

from py_figures import EIS
```

```python
# CHANGE ACCORDING TO WHERE EXAMPLE DATA IS SAVED

data_path = "C:/Users/sincl/packages/py_figures/py_figures/EIS_data/"
```

```python
# import data

expt_real, expt_imag, fit_real, fit_imag = EIS.import_cycles(data_path, "In7Zr3_cell5_100uA_03_PEIS_C01", 28)
vals = EIS.import_param_vals(data_path, "In7Zr3_cell5_100uA_03_PEIS_C01", 28)
```

# Plot a single EIS cycle

```python
title = r"Li$_{2.7}$In$_{0.7}$Zr$_{0.3}$Cl$_6$"

# plot EIS cycle 1
ax1 = EIS.plot_singlecycle(expt_real[0], expt_imag[0], fit_real[0], fit_imag[0], "#7D83FF", (0, 6500))

# get parameter labels in scientific notation
labels = EIS.labels_list([vals[0][1], vals[0][2], vals[0][3], vals[0][4], vals[0][5], vals[0][6]])

# put parameters labels in equation form
val_list = [(labels[0], "R", 1), (labels[1], "n", 1), (labels[2], "Q", 1), 
           (labels[3], "R", 2), (labels[4], "n", 2), (labels[5], "Q", 2)]

# put parameter labels in list
params = EIS.params_list(val_list)

# plot box with parameter values
EIS.plot_fit_params(ax1, params, (300, 5700))

# add title
ax1.text(200, 6300, title, ha="left", va="top", fontsize="16")

# add cycle info
txt = EIS.singlecycle_info(100, 1, 28)
ax1.text(6300, 5400, txt, ha="right", va="top", fontsize="16")
```

# Plot multiple EIS cycles

```python
# plot cycles
ax2 = EIS.plot_multicycle(28, expt_real, expt_imag, fit_real, fit_imag, (0, 12500))

# add title
ax2.text(200, 12200, title, ha="left", va="top", fontsize="16")

# add cycle info
txt = EIS.expt_info(100, 28)
ax2.text(12200, 12200, txt, ha="right", va="top", fontsize="16")
```

# Plot R values vs. cycle number

```python
# create list of R values from set of fit parameter data
R1_vals = []
for i in range(28):
    R1_vals.append(vals[i][1])

# plot R values
ax3 = EIS.plot_R(28, R1_vals, x_lim=(-0.9, 28.9), y_lim=(500, 620), marker="o")
```
